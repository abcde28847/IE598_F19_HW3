```python
#List 1 Sizing Up a New Data Set and input the datab
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import sys
```

```python
sns.set()
HY = pd.read_csv("HY_Universe_corporate bond.csv",header = 0)
HY.head()
```

```python
# List2 Determining the Nature of Attributes
print("Num of rows: " + str(HY.shape[0]))
print("Num of columns: " + str(HY.shape[1]))
HY.info()
```

```python
# List 3 Summary Statistics for Numeric and Categorical Attributes

Industry = HY['Industry']
unique = set(Industry)
sys.stdout.write("Unique Label Values \n")
print(unique) 
count = {}

for num in Industry:
    if num in count:
        count[num] = count[num]+1
    else:
        count[num] = 1   
sys.stdout.write("\nCounts for Each Value of Categorical Label \n")
print(count)
```

```python
# Listing 4: Quantile‐Quantile Plot
from scipy import stats
stats.probplot(HY.Coupon, dist="norm", plot=plt)
plt.show()
stats.probplot(HY["Issued Amount"], dist="norm", plot=plt)
plt.show()
stats.probplot(HY["LIQ SCORE"], dist="norm", plot=plt)
plt.show()
stats.probplot(HY.n_trades, dist="norm", plot=plt)
plt.show()
stats.probplot(HY["weekly_mean_volume"], dist="norm", plot=plt)
plt.show()
stats.probplot(HY["weekly_median_volume"], dist="norm", plot=plt)
plt.show()
stats.probplot(HY["weekly_max_volume"], dist="norm", plot=plt)
plt.show()
```

```python
# List 5 Using Python Pandas to Read and Summarize Data
from pandas import DataFrame
import matplotlib.pyplot as plot
print(HY.head())
print(HY.tail())
summary = HY.describe()
print(summary)
```

```python
# Listing 6: Parallel Coordinates Graph for Real Attribute Visualization
from pandas.plotting import parallel_coordinates
data = HY[['Coupon','LIQ SCORE','total_median_size','weekly_mean_volume','total_mean_size','weekly_median_volume']][:100]
parallel_coordinates(data, 'Coupon')
plt.legend(loc = 'upper right', bbox_to_anchor = (0.6,-0.1),  ncol = 2, fancybox = True, shadow = True)
plt.show()
```

```python
# Listing 7: Cross Plotting Pairs of Attributes
plt.scatter(HY["n_trades"], HY["LIQ SCORE"])
plt.xlabel('n_trades')
plt.ylabel('LIQ SCORE')
plt.show()

plt.scatter(HY["n_trades"], HY["n_days_trade"])
plt.xlabel('n_trades')
plt.ylabel('n_days_trade')
plt.show()
```

```python
# Listing 8: Correlation between Classification Target and Real Attributes
correlations = pd.DataFrame(HY.corr())
print(correlations)
```

```python
# Listing 9: Pearson’s Correlation Calculation
sys.stdout.write("Correlation between n_trades and n_days_trade \n")
np.corrcoef(HY['n_trades'], HY['n_days_trade'])[0, 1]
```

```python
# Listing 10: Presenting Attribute Correlations Visually
corMat = pd.DataFrame(HY.corr())
plot.pcolor(corMat)
plot.show()
```

```python
# Box-Plot
sns.set()
sns.boxplot(HY['Coupon'][:30], HY['n_days_trade'][:20], data=HY)
plt.show()
```

```python
# ECDF
x = np.sort(HY["LIQ SCORE"])
y = np.arange(1, len(x)+1) / len(x) 
plt.plot(x, y, marker='.', linestyle='none')
plt.xlabel("LIQ SCORE")
plt.ylabel("ECDF")
plt.show()
```
